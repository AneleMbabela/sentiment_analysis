```python

import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
reviews = pd.read_csv("Musical_instruments_reviews.csv")
```

```python
from transformers import pipeline
import torch
```

```python
pipe = pipeline('sentiment-analysis', model='arpanghoshal/EkmanClassifier',device = 0) 
```

```python
pipe("I loved the movie but the ending was shocking")
```

```python
reviews.head(10)
```

```python
def ekman_classifier(text):
    emotion = pipe(text)[0]['label']
    return(emotion)
```

```python
ekman_classifier("I'm so mad at my people")
```

```python
reviews_head = reviews.head(20)
```

```python
reviews_head['emotion'] = reviews_head['reviewText'].apply(ekman_classifier)
```
